```python
#this defines a variable "s" which is a string variable. We are then having python print the indexed character from said sequence.
# So for s[0], this gives us the first character in the word 'spam' which is s. For s[-2], this gives us the third character in the word spam (this is because python allows negative indexing which indexes from left to right and starts with -1)
#(so the character indexed -1 is 'm', the character indexed -2 is 'a', and so on)
s = 'spam'
# SW: note that the string 'spam' is assigned to the variable s
print(s[0], s[-2])

#with the command print(s[i,k]) we would be starting with the ith character and ending with the kth
# so our first example gives us "pa" since we are starting with the character indexed 1 (which is p) and ending
#with the character indexed 3 (which is m) so this just prints pa
#similarly, for s[1:] we are starting with the character indexed one again, but this time we are not designating an end
#so our result is "pam"
#for the final example, we are not designating a starting point (so we start at the 0th term) and we are ending at the term indexed -1
#which is the same as the term indexed 3(which is m), so the result is "spa"
print(s[1:3], s[1:], s[:-1])




```

```python
#here  we are importing the urllib.request module, which defines functions and classes which are helpful in opening urls
import urllib.request
#this locates and gets the course schedule from github and assigns it to the variable 'r'
r = urllib.request.urlopen('https://github.com/Math510/Math510.github.io/raw/master/2015%20FALL%20510%20Course%20Schedule.pdf')

#here we are creating a file object using the open function and assigning it to the variable 'myfile'; (the first part designating the location of the file)
#the second part is write and binary
myfile = open('/Users/harrisonkatz/Downloads/Schedule.pdf', 'wb')

#here we are applying the function write to the variable 'myfile' and whatever comes in the parathensis of write()
#gets written onto the variable 'myfile', so for our case whatever is read on 'r' gets written onto 'myfile'
myfile.write(r.read())

#after the file content is written onto myfile, '.close()' is used to close the file object (very important!)
myfile.close()
```
